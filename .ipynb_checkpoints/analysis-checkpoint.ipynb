```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture as GM
import geocoder
```

```python
df = pd.read_csv('ssh-logs.csv')
df.time = pd.to_datetime(df.time, format='%Y-%m-%dT%H:%M:%S', utc=True)
```

observation: if fingerprint is present then password is NaN

```python
def f(group):
    group = group.sort_values('time')
    t = group.time.values.astype('datetime64[s]')
    dt = np.zeros(len(group))
    dt[1:] = pd.Series(t[1:] - t[:-1]).dt.seconds
    dt[0] = None
    return pd.Series(dt, index=group.index)

grouped = df.groupby('src').apply(f)
grouped
```

```python
dt = grouped.reset_index(level='src')[0].rename('dt')
df = df.join(dt)
df
```

```python
df.src.value_counts()
```

```python
dt = df[df.src == '221.194.44.redacted'].dt
np.log(dt[dt>0]).hist()
```

```python
dt = df.set_index('src', append=True).dt.fillna(0)

lg_dt = np.log(dt + .0001)
lg_dt_agg = lg_dt.groupby('src').agg(['mean', 'std']).fillna(0)
lg_dt_agg.columns = [f'dt_{col}' for col in lg_dt_agg.columns]

kmeans = KMeans(n_clusters=4).fit(lg_dt_agg)
lg_dt_agg['label'] = kmeans.labels_
lg_dt_agg
```

```python
plt.scatter(lg_dt_agg.dt_mean, lg_dt_agg.dt_std, c=lg_dt_agg.label)
plt.xlabel('mean')
plt.ylabel('std')
```

```python
dt = df.set_index('src', append=True).dt.fillna(0)

lg_dt = np.log(dt + .0001)
lg_dt_agg = lg_dt.groupby('src').agg(['mean', 'std']).fillna(0)
lg_dt_agg.columns = [f'dt_{col}' for col in lg_dt_agg.columns]

gm = GM(n_components=3).fit(lg_dt_agg)
lg_dt_agg['label'] = gm.predict(lg_dt_agg)
lg_dt_agg
```

```python
plt.scatter(lg_dt_agg.dt_mean, lg_dt_agg.dt_std, c=lg_dt_agg.label)
plt.xlabel('mean')
plt.ylabel('std')
```
